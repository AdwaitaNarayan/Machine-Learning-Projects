```python
import matplotlib.pyplot as plt
import plotly.express as px
```

### Analyzing data uisng bar graph

```python
grouped = df.groupby('Region')[' Estimated Unemployment Rate (%)'].mean().reset_index()
grouped = grouped.sort_values(by=' Estimated Unemployment Rate (%)', ascending=False)
regions = grouped['Region']
unemployment_rates = grouped[' Estimated Unemployment Rate (%)']
plt.style.use('dark_background')
plt.figure(figsize=(10, 6))
plt.bar(regions, unemployment_rates, color='red')
plt.title('Unemployment Rate by Region')
plt.xlabel('Region -->')
plt.ylabel(' Estimated Unemployment Rate (%) -->')
plt.xticks(rotation=90)

plt.tight_layout()

plt.show()
```

### Analyzing data uisng box plot

```python
plt.figure(figsize=(10, 6))
boxplot = df.boxplot(column=' Estimated Unemployment Rate (%)', by='Region', grid=False, rot=90)

plt.title('Unemployment Rate by Region (Box Plot)')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.show()
```

### Analyzing data uisng scatter plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Region'].astype(str), df[' Estimated Unemployment Rate (%)'], color = 'red')
plt.title('Unemployment Rate by Region (Scatter Plot)')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### Analyzing data uisng histogram

```python
plt.figure(figsize=(14, 8))
plt.hist([df[df['Region'] == region][' Estimated Unemployment Rate (%)'] for region in df['Region']], bins=10, alpha=0.5, label=df['Region'].unique())
plt.title('Unemployment Rate by Region (Histogram)')
plt.xlabel('Estimated Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.legend(title='Region')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
